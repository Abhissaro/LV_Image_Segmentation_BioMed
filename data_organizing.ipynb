```python
def makedirs():

  os.makedirs('data/original_data', exist_ok=True)
  shutil.move('/content/drive/MyDrive/LVEF/CAMUS/data/database', '/content/drive/MyDrive/LVEF/CAMUS/original_data')

  os.makedirs('data/testing/2ch/frames', exist_ok=True)
  os.makedirs('data/testing/4ch/frames', exist_ok=True)
  os.makedirs('data/testing/2ch/masks', exist_ok=True)
  os.makedirs('data/testing/4ch/masks', exist_ok=True)
  os.makedirs('data/testing/2ch/sequences', exist_ok=True)
  os.makedirs('data/testing/4ch/sequences', exist_ok=True)
  os.makedirs('data/testing/2ch/info', exist_ok=True)
  os.makedirs('data/testing/4ch/info', exist_ok=True)

  return 
```

### Reorganize  Training Files

```python
def reorganize_tng_file():

  for _dir, _, files in os.walk('/content/drive/MyDrive/LVEF/CAMUS/original_data/training'):
      for file in files:
          if file.endswith('.mhd') or file.endswith('.raw'):
              print(os.path.join(_dir, file))
              if '2CH' in file:
                  if 'gt' in file:
                      shutil.move(os.path.join(_dir, file), '/content/drive/MyDrive/LVEF/CAMUS/original_data/data/training/2ch/masks/')
                  elif 'sequence' in file:
                      shutil.move(os.path.join(_dir, file), '/content/drive/MyDrive/LVEF/CAMUS/original_data/data/training/2ch/sequences/')
                  else:
                      shutil.move(os.path.join(_dir, file), '/content/drive/MyDrive/LVEF/CAMUS/original_data/data/training/2ch/frames/')
              else:
                  if 'gt' in file:
                      shutil.move(os.path.join(_dir, file), '/content/drive/MyDrive/LVEF/CAMUS/original_data/data/training/4ch/masks/')
                  elif 'sequence' in file:
                      shutil.move(os.path.join(_dir, file), '/content/drive/MyDrive/LVEF/CAMUS/original_data/data/training/4ch/sequences/')
                  else:
                      shutil.move(os.path.join(_dir, file), '/content/drive/MyDrive/LVEF/CAMUS/original_data/data/training/4ch/frames/')
          elif file.endswith('.cfg'):
              patient = os.path.basename(_dir)
              if '2CH' in file:
                  shutil.move(os.path.join(_dir, file), f'/content/drive/MyDrive/LVEF/CAMUS/original_data/data/training/2ch/info/{patient}_info.cfg')
              else:
                  shutil.move(os.path.join(_dir, file), f'/content/drive/MyDrive/LVEF/CAMUS/original_data/data/training/4ch/info/{patient}_info.cfg')

  return 
```
